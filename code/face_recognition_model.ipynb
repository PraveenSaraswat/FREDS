```
from os import listdir
import face_recognition
import os,cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import color
from skimage import io
import pickle
from sklearn import svm
```

```
# Created function which is calculating representation for the employee images stored in the database

def emp_representation(emp_pic_path):
  train_dir = os.listdir(emp_pic_path)
  for person in train_dir: 
    pix = os.listdir(emp_pic_path + person)
    for person_img in pix:
      face = face_recognition.load_image_file(emp_pic_path + person + "/" + person_img)
      face_bounding_boxes = face_recognition.face_locations(face)
  # If training image contains exactly one face 
      if len(face_bounding_boxes) == 1: 
        face_enc = face_recognition.face_encodings(face, num_jitters=100)[0] 
                # Add face encoding for current image  
                # with corresponding label (name) to the training data 
        train_data_encodings.append(face_enc) 
        train_data_names.append(person) 
      else: 
        print(person + "/" + person_img + " can't be used for training")      
	
  print("employee representations retrieved successfully")
```

```
employee_pictures = "/content/drive/My Drive/Colab Notebooks/Capstone Project/CP_Final/Database/"
model_save_path = '/content/drive/My Drive/Colab Notebooks/Capstone Project/CP_Final/Weights/face_recog.sav'

final_names =[]

train_data_encodings = [] 
train_data_names = []
```

```
# Populating the employees dictionary with the representation of the employees in database.
emp_representation(employee_pictures)
clf = svm.SVC(gamma ='scale', probability =True) 
clf.fit(train_data_encodings, train_data_names)
```

```

if model_save_path is not None:
  with open(model_save_path, 'wb') as f:
    pickle.dump(clf, f)
```
